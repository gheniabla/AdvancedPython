# Chapter 24 Four Pillars of Object Oriented Programming

Abstraction, Encapsulation, Inheritance, and Polymorphism are the four pillars of object-oriented programming.

## Chapter 24.1 Abstraction

Abstraction is the process of hiding complex implementation details and exposing only the essential features of an object. It lets you focus on what an object does, not how it does it.
Example: An abstract Animal class defines a make_sound() method without specifying how each animal sounds.


```python
from abc import ABC, abstractmethod

class Animal(ABC):
    @abstractmethod
    def make_sound(self):
        pass
```

## Chapter 24.2 Encapsulation

Encapsulation means keeping the data (attributes) private inside a class, and exposing it only through controlled methods (getters/setters).
Example: A Human class with private fields and public methods to access them:

```python
class Human:
    def __init__(self, name, age):
        self.__name = name  # private attribute
        self.__age = age    # private attribute

    def get_name(self):
        return self.__name

    def set_name(self, name):
        self.__name = name

    def get_age(self):
        return self.__age

    def set_age(self, age):
        if age >= 0:
            self.__age = age
        else:
            print("Age must be non-negative")
```

## Chapter 24.3 Inheritance

Inheritance allows a class to inherit attributes and methods from a parent class, promoting reuse and hierarchy.
Example: Cat, Dog, and Bird inherit from Animal:

```python
class Cat(Animal):
    def make_sound(self):
        print("Meow")

class Dog(Animal):
    def make_sound(self):
        print("Bark")

class Bird(Animal):
    def make_sound(self):
        print("Chirp")
```

## Chapter 24.4 Polymorphism

Polymorphism allows different classes to be treated through a common interface, even though they behave differently.
Example: A list of Animal objects where each one behaves according to its actual type:

```python
def animal_sound(animal):
    animal.make_sound()

animals = [Cat(), Dog(), Bird()]

for animal in animals:
    animal_sound(animal)
```
